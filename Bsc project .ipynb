# 2. Loading the data and finding out basic information

```python
# reading data
sales = pd.read_excel("Sales_project (1).xlsx")
customers=pd.read_excel('Sales_project (1).xlsx', 'Customers Sheet')
sites = pd.read_excel('Sales_project (1).xlsx', 'Site Location sheet')
products=pd.read_excel('Sales_project (1).xlsx', 'Products_Sheet')
sales_team= pd.read_excel('Sales_project (1).xlsx','Sales Team Sheet')
```

```python
# splliting order date column into separate year, monthh and day columns 
```

```python
year = []
month = []
day = []
for date in sales['OrderDate']:
    temp = str(date).split(" ")[0].split("-")
    year.append(temp[0])
    month.append(temp[1])
    day.append(temp[2])

sales['Year'] = year
sales['Month'] = month
sales['Day'] = day
sales.head()
```

```python
#Merge the sales and customers and site # they much common columns
data_merged = sales.join(customers.set_index('_CustomerID'), on='_CustomerID').join(sites.set_index('_SiteID'), on='_SiteID').join(products, on='_ProductID', how='left', lsuffix='_left', rsuffix='_right')
```

# 3. Cleaning/ adding relevant and Dropping irrelevant columns and entries

```python
# adding new variable
data_merged['total_sales']= data_merged['Unit Cost'] * data_merged['Unit Price']
```

```python
# adding new varibale
data_merged['total_sales'] = data_merged['_SalesAM_ID']/data_merged['_CustomerID']
data_merged['total_sales'].describe()
```

```python
# checking for null values
data_merged.isnull().sum()
```

```python
# cleaning data 
#fill missing values by average of a column
data_merged = data_merged.apply(lambda x:x.fillna(x.value_counts().index[0]))
```

```python
#Sometimes there would be many columns that we never use in such cases dropping is the only solution.
data_merged.drop(['OrderDate', 'CurrencyCode'], axis=1, inplace=True)
```
